Question 1: Find an approximate solution to the following using Gradient descent.

$$min$$ 
$$ x^2 + 3x + 6$$
1. Choose the learning rate as 10−1 , plot objective function value for each iteration and report your observations.
2. Now choose the learning rate as 10−4, plot objective function value for each iteration and report your observations.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 3*x + 6

def grad_f(x):
    return 2*x + 3

def gradient_descent(learning_rate, initial_x, num_iterations):
    x = initial_x
    values = []
    for i in range(num_iterations):
        x = x - learning_rate * grad_f(x)
        values.append(f(x))
    return values

initial_x = 0
num_iterations = 100

learning_rate_1 = 1e-1
learning_rate_2 = 1e-4

values_1 = gradient_descent(learning_rate_1, initial_x, num_iterations)
values_2 = gradient_descent(learning_rate_2, initial_x, num_iterations)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(values_1, label='Learning rate = 1e-1')
plt.xlabel('Iteration')
plt.ylabel('Objective function value')
plt.title('Gradient Descent with Learning Rate 1e-1')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(values_2, label='Learning rate = 1e-4')
plt.xlabel('Iteration')
plt.ylabel('Objective function value')
plt.title('Gradient Descent with Learning Rate 1e-4')
plt.legend()

plt.tight_layout()
plt.show()
print(f'learning rate 1e-1: {values_1[-1]}')
print(f'learning rate 1e-4: {values_2[-1]}')
```

Question 2: Find the solution to the following using gradient descent. You can
choose an appropriate learning rate like {10−1, 10−2, 10−3, 10−4} etc...
1. min $$\frac{1}{2}x^3+5x^2 + x + 4$$ 
   1. st − 2 ≤ x ≤ 2

2.  min $$x^{2}_{1} +x^{2}_{2} + 4x_2 + 3x_1 +7 $$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 0.5 * x**3 + 5 * x**2 + x + 4

def grad_f(x):
    return 1.5 * x**2 + 10 * x + 1

def gradient_descent(learning_rate, initial_x, num_iterations, lower_bound, upper_bound):
    x = initial_x
    values = []
    for i in range(num_iterations):
        x = x - learning_rate * grad_f(x)
        x = np.clip(x, lower_bound, upper_bound)
        values.append(f(x))
    return values

initial_x = 0
num_iterations = 100
lower_bound = -2
upper_bound = 2

learning_rates = [1e-1, 1e-2, 1e-3, 1e-4]
results = {}

for lr in learning_rates:
    values = gradient_descent(lr, initial_x, num_iterations, lower_bound, upper_bound)
    results[lr] = values

plt.figure(figsize=(14, 10))

for lr, values in results.items():
    plt.plot(values, label=f'Learning rate = {lr}')

plt.xlabel('Iteration')
plt.ylabel('Objective function value')
plt.title('Gradient Descent with Different Learning Rates')
plt.legend()
plt.show()
print(f'learning rate 1e-1: {results[1e-1][-1]}')
print(f'learning rate 1e-2: {results[1e-2][-1]}')
print(f'learning rate 1e-3: {results[1e-3][-1]}')
print(f'learning rate 1e-4: {results[1e-4][-1]}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2):
    return x1**2 + x2**2 + 4*x2 + 3*x1 + 7

def grad_f(x1, x2):
    df_dx1 = 2*x1 + 3
    df_dx2 = 2*x2 + 4
    return np.array([df_dx1, df_dx2])

def gradient_descent_2d(learning_rate, initial_x, num_iterations):
    x = np.array(initial_x)
    values = []
    for i in range(num_iterations):
        grad = grad_f(x[0], x[1])
        x = x - learning_rate * grad
        values.append(f(x[0], x[1]))
    return values

initial_x = [0, 0]
num_iterations = 100

learning_rates = [1e-1, 1e-2, 1e-3, 1e-4]
results = {}

for lr in learning_rates:
    values = gradient_descent_2d(lr, initial_x, num_iterations)
    results[lr] = values

plt.figure(figsize=(14, 10))

for lr, values in results.items():
    plt.plot(values, label=f'Learning rate = {lr}')

plt.xlabel('Iteration')
plt.ylabel('Objective function value')
plt.title('Gradient Descent for Problem 2 with Different Learning Rates')
plt.legend()
plt.show()

for lr in learning_rates:
    print(f'Learning rate {lr}: {results[lr][-1]}')
```
